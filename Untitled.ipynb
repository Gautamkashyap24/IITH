```python
x = D1[['Time_Minutes','Motor_Speed_RPM','Compression_Force_kN','Flow_Rate_LPM','Vibration_mm_s','Temperature_C','Pressure_Bar','Humidity_Percent']]
```

```python
y = D1['Power_Consumption_kW']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x, y , test_size = 0.2 , random_state=100)
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor

R = RandomForestRegressor(max_depth=2, random_state=100)
R.fit(x_train, y_train)
```

```python
y_r_train_predict = R.predict(x_train)
y_r_test_predict = R.predict(x_test)
```

```python
y_r_train_predict 
```

```python
y_r_test_predict 
```

```python
from sklearn.metrics import mean_squared_error, r2_score
R_train_mse = mean_squared_error(y_train , y_r_train_predict)
R_train_r2 = r2_score(y_train, y_r_train_predict)

R_test_mse = mean_squared_error(y_test, y_r_test_predict)
R_test_r2 = r2_score(y_test, y_r_test_predict)

import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
R_train_rmse = np.sqrt(mean_squared_error(y_train, y_r_train_predict))
R_test_rmse = np.sqrt(mean_squared_error(y_test, y_r_test_predict))

R_train_mape = mean_absolute_percentage_error(y_train, y_r_train_predict) * 100
R_test_mape = mean_absolute_percentage_error(y_test, y_r_test_predict) * 100
```

```python
print('Train MSE of Random Forest:',R_train_mse)
print('Train R2 of Random Forest:',R_train_r2)
print('Train RMSE of Random Forest:',R_train_rmse)
print('Train MAPE of Random Forset:',R_train_mape)
print('Test MSE of Random Forest:',R_test_mse)
print('Test R2 of Random Forest:',R_test_r2)
print('Test RMSE of Random Forest:',R_test_rmse)
print('Test MAPE of Random Forest:',R_test_mape)
```
