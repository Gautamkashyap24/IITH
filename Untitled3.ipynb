```python
import pandas as pd
import numpy as np
import xgboost as xgb
import random
import re
```

 # GENERATE SYNTHETIC RETAIL DATASET

```python
def generate_personal_loan_data(num_samples=2000):
    """
    Creates a dummy dataset of individuals to train the ML model.
    The model needs to learn how CIBIL, Passbook Balance, and ITR affect eligibility.
    """
    np.random.seed(42)
    
    # Generate realistic ranges for individual applicants
    cibil_scores = np.random.randint(300, 900, num_samples)
    requested_amounts = np.random.uniform(50000, 1500000, num_samples) # ₹50k to ₹15L
    passbook_avg_balances = np.random.uniform(5000, 200000, num_samples)
    
    # ITR is optional, so we'll simulate some having it (1) and some not (0)
    has_itr = np.random.choice([0, 1], size=num_samples, p=[0.4, 0.6])
    itr_incomes = has_itr * np.random.uniform(300000, 2500000, num_samples) 
    
    df = pd.DataFrame({
        'cibil_score': cibil_scores,
        'requested_amount': requested_amounts,
        'passbook_avg_balance': passbook_avg_balances,
        'has_itr': has_itr,
        'itr_income': itr_incomes
    })
    
    # Define Target Logic (1 = Reject, 0 = Approve)
    # The AI will learn these patterns from the data
    conditions = (
        (df['cibil_score'] < 650) | # Low CIBIL is an instant red flag
        (df['requested_amount'] > (df['passbook_avg_balance'] * 20)) | # Asking for way more than they hold
        ((df['has_itr'] == 0) & (df['requested_amount'] > 500000)) # Asking for >5L without an ITR
    )
    df['risk_label'] = np.where(conditions, 1, 0)
    
    return df
```

# THE RETAIL LOAN ENGINE (Inference)

```python
class PersonalLoanEngine:
    def __init__(self):
        print("Training AI on dummy retail datasets...")
        self.model = self._train_model()
        print("Model trained successfully!\n")

    def _train_model(self):
        """Trains the XGBoost model on the generated dummy data."""
        df = generate_personal_loan_data()
        X = df[['cibil_score', 'requested_amount', 'passbook_avg_balance', 'has_itr', 'itr_income']]
        y = df['risk_label']
        
        model = xgb.XGBClassifier(n_estimators=50, max_depth=4, random_state=42)
        model.fit(X, y)
        return model

    # --- MOCK APIs AND EXTRACTION ---
    def fetch_cibil_score(self, pan_number):
        """Mocks an API call to a credit bureau using the PAN number."""
        # In real life, use Python `requests.get()` to hit the CIBIL API
        # Here we generate a consistent mock score based on the PAN string length/chars
        random.seed(pan_number)
        score = random.randint(550, 850)
        print(f"[API Call] Fetched CIBIL Score for {pan_number}: {score}")
        return score

    def extract_passbook_data(self, image_path):
        """Mocks OCR extraction from an uploaded passbook image."""
        # In real life, use EasyOCR or pytesseract here.
        print(f"[OCR] Scanning {image_path} for transaction history...")
        # Mocking the extracted average monthly balance
        mock_extracted_balance = random.uniform(25000, 150000) 
        print(f"[OCR] Extracted Average Monthly Balance: ₹{mock_extracted_balance:,.2f}")
        return mock_extracted_balance

    # --- FINAL EVALUATION ---
    def evaluate_applicant(self, name, pan, requested_amount, passbook_image, itr_income=0):
        """Runs the full pipeline for a single user."""
        print(f"--- Processing Application for {name} ---")
        
        # 1. Fetch live data
        cibil_score = self.fetch_cibil_score(pan)
        passbook_balance = self.extract_passbook_data(passbook_image)
        has_itr = 1 if itr_income > 0 else 0
        
        # 2. Build the feature vector
        features = pd.DataFrame([{
            'cibil_score': cibil_score,
            'requested_amount': requested_amount,
            'passbook_avg_balance': passbook_balance,
            'has_itr': has_itr,
            'itr_income': itr_income
        }])
        
        # 3. Predict using the trained model
        risk_prediction = self.model.predict(features)[0]
        risk_probability = self.model.predict_proba(features)[0][1]
        
        # 4. Generate Decision Logic
        if risk_prediction == 1 or risk_probability > 0.6:
            status = "REJECTED"
            approved_limit = 0
            rate = 0.0
            explanation = "High risk profile based on CIBIL or insufficient cash flow."
        else:
            status = "APPROVED"
            # Cap the limit based on passbook balance and CIBIL tier
            approved_limit = min(requested_amount, passbook_balance * 15)
            # Better CIBIL gets a lower interest rate
            rate = max(9.5, 18.0 - (cibil_score / 100))
            explanation = "Acceptable credit history and sufficient account balance."

        print(f"\n--- DECISION ---")
        print(f"Status: {status}")
        if status == "APPROVED":
            print(f"Approved Limit: ₹{approved_limit:,.2f}")
            print(f"Interest Rate: {rate:.2f}% p.a.")
        print(f"Reason: {explanation}")
        print("-" * 40)
```

# RUN THE PIPELINE

```python
if __name__ == "__main__":
    engine = PersonalLoanEngine()
    
    # Applicant 1: Strong applicant with ITR
    engine.evaluate_applicant(
        name="Amit Kumar",
        pan="ABCDE1234F",
        requested_amount=300000,
        passbook_image="amit_sbi_passbook.jpg",
        itr_income=800000 # Provided ITR
    )
    
    # Applicant 2: Risky applicant (Asking for too much without ITR)
    engine.evaluate_applicant(
        name="Rahul Sharma",
        pan="ZYXWV9876Q",
        requested_amount=800000,
        passbook_image="rahul_hdfc_passbook.jpg",
        itr_income=0 # Did not provide ITR
    )
```
